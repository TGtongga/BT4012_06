```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv('/content/drive/My Drive/BT4012 Group 06/Data/train_non_embedded.csv', index_col=None)
test = pd.read_csv('/content/drive/My Drive/BT4012 Group 06/Data/test_non_embedded.csv', index_col=None)
X_train = train.drop('fraudulent',axis = 1)
y_train = train[['fraudulent']]
X_test = test.drop('fraudulent',axis = 1)
y_test = test[['fraudulent']]
```

Linear SVM

```python
model1 = SVC(kernel='linear', C=1.0)
model1.fit(X_train, y_train)
```

```python
y_pred1 = model1.predict(X_test)
accuracy1 = accuracy_score(y_test, y_pred1)
accuracy1
```

```python
feature_coefficients = model1.coef_
feature_coefficients
```

```python
len(feature_coefficients[0])
```

```python
abs_feature_coefficients = np.abs(feature_coefficients)

sorted_indices = np.argsort(abs_feature_coefficients, axis=None)[::-1]

top_10_indices1 = sorted_indices[:10]

top_10_columns1 = train.columns[top_10_indices1 ]

top_10_coefficient1= feature_coefficients[0][top_10_indices1]
```

```python
print(top_10_columns1)
print(top_10_coefficient1)
```

Polynomial

```python
model2 = SVC(kernel='poly', C=1.0)
model2.fit(X_train, y_train)
```

```python
y_pred2 = model2.predict(X_test)
accuracy2 = accuracy_score(y_test, y_pred2)
accuracy2
```

```python
support_vectors = model2.support_vectors_
dual_coefficients = model2.dual_coef_
```

```python
support_vectors
```

```python
dual_coefficients
```

```python
max_dual_coefficient = np.max(model2.dual_coef_)
max_dual_coefficient_index = np.argmax(model2.dual_coef_)
corresponding_support_vector = model2.support_vectors_[max_dual_coefficient_index]
```

```python
corresponding_support_vector
```

```python
abs_corresponding_support_vector = np.abs(corresponding_support_vector)

sorted_indices = np.argsort(abs_corresponding_support_vector, axis=None)[::-1]

top_10_indices2 = sorted_indices[:10]

top_10_columns2 = train.columns[top_10_indices2 ]

top_10_coefficient2= corresponding_support_vector[top_10_indices2]
```

```python
print(top_10_columns2)
print(top_10_coefficient2)
```
