<h1> Model Performance Plot </h1>

This notebook is used to plot the performance of the models. The performance is measured by the F1 score on the test set.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
f1_scores = {
    'Logistic Regression': 0.80392,
    'SVM': 0.87805,
    'Random Forest': 0.8,
    'KNN': 0.8339,
    'LightGBM': 0.88811,
    'XGBoost': 0.86219,
    'Naive Bayes': 0.44216,
    'LSTM': 0.89895,
    'Rare Event Logistic': 0.79154
}
```

```python

sns.set_theme(style="whitegrid")
f1_df = pd.DataFrame.from_dict(f1_scores, orient='index')
f1_df.reset_index(inplace=True)
f1_df.columns = ['Model', 'F1 Score']
f1_df = f1_df.sort_values(by=['F1 Score'], ascending=False)
f1_df = f1_df.reset_index(drop=True)
f1_df['Rank'] = f1_df.index + 1
f1_df['F1 Score'] = f1_df['F1 Score'].round(4)
f1_df
```

```python
# Plot the F1 scores using seaborn as a bar chart
plt.figure(figsize=(10, 6))
sns.set(font_scale=1.3)
sns.barplot(x='Model', y='F1 Score', data=f1_df, palette='flare') #available palettes: https://seaborn.pydata.org/tutorial/color_palettes.html
plt.xticks(rotation=45)
plt.xlabel('Model')
plt.ylabel('F1 Score')
plt.title('F1 Score of Models of TF-IDF after Hyperparameter Tuning')
plt.tight_layout()

# Plot the exact F1 scores on top of the bars
for index, row in f1_df.iterrows():
    plt.text(row.name, row['F1 Score'] + 0.01, row['F1 Score'], color='black', ha="center", fontdict={'fontsize': 12})

# Set x-label to be smaller
plt.tick_params(axis='x', labelsize=12)

# Show the plot
plt.show()
```
